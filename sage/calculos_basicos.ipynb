## Funciones trigonométricas 

```python
var("n, p,k")
```

```python
expresion = binomial(n, k)*p^k*(1-p)^(n-k)
show(expresion)
```

```python
valor= expresion.substitute(k=0, n=10, p=1/5)+expresion.substitute(k=1, n=10, p=1/5)+expresion.substitute(k=2, n=10, p=1/5)+expresion.substitute(k=3, n=10, p=1/5)
```

```python
valor.n(digits=3)
```

```python
f(n, p, k)= binomial(n, k)*p^k*(1-p)^(n-k)
f
```

```python
val = f(10, 1/5, 0)+f(10, 1/5, 1)+f(10, 1/5, 2)+f(10, 1/5, 3)
val.n(digits=4)
```
